## uji regresi

```python
import statsmodels.api as sm

X = df[['sales', 'quantity','discount', 'shipping_cost', 'year']]
y = df[['profit']]

X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

print_model = model.summary()
print(print_model)
```

* Sales, Quantity, Shipping Cost, Discount Memiliki nilai p-value yang sangat kecil (0.000) sehingga menunjukkan bahwa penjualan memiliki pengaruh signifikan positif terhadap profit; kenaikan dalam penjualan akan meningkatkan profit secara signifikan.

## segment dengan profit

```python
profit_per_segment = df.groupby('segment')['profit'].sum()

# Membuat pie chart
plt.figure(figsize=(8, 8))
plt.pie(profit_per_segment, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Total Profit per Segment')
plt.axis('equal')  # Membuat pie chart menjadi lingkaran

# Menambahkan legend di atas kanan
plt.legend(profit_per_segment.index, title='Segment', bbox_to_anchor=(1, 1), loc='upper left')

plt.show()
```

## top 3 region dengan profit

```python
profit_per_region = df.groupby('region')['profit'].sum().nlargest(3)

# Membuat grafik batang
profit_per_region.plot(kind='bar', color='skyblue')
plt.title('Total Profit by region')
plt.xlabel('region')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
profit_per_region = df.groupby('region')['profit'].sum().nsmallest(3).sort_values(ascending=False)

# Membuat grafik batang
profit_per_region.plot(kind='bar', color='lightblue')
plt.title('Top 3 regions by Lowest Total Profit')
plt.xlabel('region')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```

## kesimpulan

* Sales (Penjualan):
    Penjualan yang rendah berhubungan dengan profit yang lebih rendah. Strategi untuk meningkatkan penjualan harus difokuskan, seperti promosi, iklan yang lebih tepat sasaran atau penawaran khusus untuk menarik pelanggan baru dan mempertahankan pelanggan yang sudah ada dengan membership.

* Quantity (Jumlah):
    Jumlah produk yang rendah menunjukkan bahwa produk masih belum/tidak menarik bagi pelanggan sehingga perlu dilakukan analisis data pelanggan untuk memahami preferensi dan kebutuhan mereka dapat membantu dalam merancang produk yang lebih sesuai dan meningkatkan volume penjualan.

* Shipping Cost (Biaya Pengiriman):
    Biaya pengiriman yang tinggi dapat mengurangi daya tarik produk, terutama untuk kategori dengan profit rendah sehingga perlu solusi untuk mengurangi biaya pengiriman, seperti bernegosiasi dengan penyedia layanan logistik atau menawarkan opsi pengiriman gratis dengan syarat tertentu.

* Discount (Diskon):
    Diskon yang tinggi pada produk dengan profit rendah dapat memperburuk situasi. Di sisi lain, diskon memang dapat meningkatkan penjualan, tetapi terlalu banyak diskon dapat menggerus margin keuntungan. Mengelola strategi diskon dengan lebih hati-hati, seperti menawarkan diskon untuk pembelian dalam jumlah besar atau untuk pelanggan loyal saja.
