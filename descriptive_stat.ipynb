```python
import pandas as pd
import numpy as np

dataset= pd.read_csv('SuperStore_Orders.csv', sep=',', encoding='ISO-8859-1')
dataset

df = pd.DataFrame(dataset)
df
```

```python
# Mengganti koma dengan titik dalam kolom 'sales'
df['sales'] = df['sales'].str.replace(',', '.', regex=False)
df['sales'] = df['sales'].astype(float)
print(df.dtypes)
```

## statistical method

```python
df.describe()
```

```python
stats = {}

# Statistik deskriptif untuk setiap kolom numerik
for column in df.select_dtypes(include='number').columns:
    mean = df[column].mean()
    median = df[column].median()
    mode = df[column].mode()[0]
    variance = df[column].var()
    std_dev = df[column].std()
    max_value = df[column].max()
    min_value = df[column].min()
    range_value = max_value - min_value
    
    stats[column] = {
        'Mean': mean,
        'Median': median,
        'Mode': mode,
        'Variance': variance,
        'Standard Deviation': std_dev,
        'Range': range_value,
        'Max': max_value,
        'Min': min_value
    }

stats_df = pd.DataFrame(stats).T
print(stats_df)
```

```python
import pandas as pd

# Inisialisasi dictionary untuk menyimpan statistik
stats = {}

# Misalkan kita memiliki kolom kategori bernama 'category'
for category, group in df.groupby('category'):
    stats[category] = {}
    
    # Menghitung statistik deskriptif untuk setiap kolom numerik
    for column in group.select_dtypes(include='number').columns:
        mean = group[column].mean()
        median = group[column].median()
        mode = group[column].mode()[0]  # Ambil nilai modus pertama
        variance = group[column].var()
        std_dev = group[column].std()
        max_value = group[column].max()
        min_value = group[column].min()
        range_value = max_value - min_value
        
        stats[category][column] = {
            'Mean': mean,
            'Median': median,
            'Mode': mode,
            'Variance': variance,
            'Standard Deviation': std_dev,
            'Range': range_value,
            'Max': max_value,
            'Min': min_value
        }

# Mengonversi dictionary menjadi DataFrame
stats_df = pd.concat({k: pd.DataFrame(v) for k, v in stats.items()})

# Menampilkan DataFrame statistik
print(stats_df)
```

```python
import pandas as pd

# Inisialisasi dictionary untuk menyimpan statistik
stats = {}

# Misalkan kita memiliki kolom kategori bernama 'segment'
for segment, group in df.groupby('segment'):
    stats[segment] = {}
    
    # Menghitung statistik deskriptif untuk setiap kolom numerik
    for column in group.select_dtypes(include='number').columns:
        mean = group[column].mean()
        median = group[column].median()
        mode = group[column].mode()[0] 
        variance = group[column].var()
        std_dev = group[column].std()
        max_value = group[column].max()
        min_value = group[column].min()
        range_value = max_value - min_value
        
        stats[segment][column] = {
            'Mean': mean,
            'Median': median,
            'Mode': mode,
            'Variance': variance,
            'Standard Deviation': std_dev,
            'Range': range_value,
            'Max': max_value,
            'Min': min_value
        }

# Mengonversi dictionary menjadi DataFrame
stats_df = pd.concat({k: pd.DataFrame(v) for k, v in stats.items()})

# Menampilkan DataFrame statistik
print(stats_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

if 'discount' in df.columns:
    # Mendefinisikan bin dan label untuk diskon dalam format desimal
    bins = [0.00, 0.25, 0.50, 0.75, 1.00]
    labels = ['0-25%', '25-50%', '50-75%', '75-100%']

    # Mengelompokkan discount ke dalam rentang
    df['discount_range'] = pd.cut(df['discount'], bins=bins, labels=labels, right=True)

    # Menghitung frekuensi untuk setiap rentang
    distribution = df['discount_range'].value_counts().sort_index()

    # Membuat pie chart dengan persentase
    plt.figure(figsize=(8, 8))
    wedges, texts, autotexts = plt.pie(distribution, labels=None, autopct='%1.1f%%', startangle=140)
    
    # Menambahkan legend
    plt.legend(wedges, distribution.index, title="Discount Range", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    
    plt.title('Distribusi Diskon dalam Bentuk Pie Chart')
    plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
    
    # Mengatur ukuran font untuk persentase
    for text in autotexts:
        text.set_color('white')  # Mengatur warna teks persentase

    plt.show()
else:
    print("Kolom 'discount' tidak ditemukan dalam dataset.")
```

### shipping_cost

```python
if 'shipping_cost' in df.columns and 'category' in df.columns:
    # Mengelompokkan data berdasarkan kategori dan shipping_cost
    df['shipping_cost_range'] = df['shipping_cost'].apply(lambda x: '<=26' if x <= 26 else '>26')

    # Menghitung frekuensi untuk setiap kategori dan rentang shipping_cost
    distribution = df.groupby(['category', 'shipping_cost_range']).size().unstack(fill_value=0)

    # Menampilkan distribusi
    print(distribution)
else:
    print("Kolom 'shipping_cost' atau 'category' tidak ditemukan dalam dataset.")
```

```python
# Menghitung total untuk setiap kategori
total_counts = distribution.sum(axis=1)

# Menghitung persentase
percentage_distribution = distribution.div(total_counts, axis=0) * 100

# Menghitung rasio
ratio_distribution = distribution['<=26'] / distribution['>26']

# Menampilkan hasil
print("Persentase Distribusi:\n", percentage_distribution)
print("\nRasio <=26 terhadap >26:\n", ratio_distribution)
```

```python
# Menghitung rata-rata shipping_cost per category
average_shipping_cost = df.groupby('category')['shipping_cost'].mean()

# Menampilkan hasil
print("Rata-rata Biaya Pengiriman per Kategori:\n", average_shipping_cost)
```
